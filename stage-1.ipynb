```python
li = ['created_at','id_str','text','user','place','geo','coordinates']
```

```python
#creating a dictionary in which we are going to save ourtweets in json format

l =[]
for status in searched_tweets:
    
    json_str = json.dumps(status._json)

    parsed = json.loads(json_str)

    json.dumps(parsed, indent=4, sort_keys=True)
    lst = li
    
    l.append(dict((k, parsed[k]) for k in lst))
    
with open('data_file.json', 'w+') as f:          #writing file in jsnn format
    json.dump(l, f, indent=4)
```
